```python
import matplotlib.pyplot as plt
```

```python
# First values for x1 and x2
x1_current = 1
x2_current = 5

# Defining lists for x1 and x2 to store the value of these two variables in
x1 = [x1_current]
x2 = [x2_current]

p1 = -4
p2 = 4
```

```python

for i in range(2,51):
  
  # First, calculating c1 and c2 based on the current x1 and x2
  c1_current = x1_current + (x1_current * x2_current) 
  c2_current = x2_current + 1

  # Calculating x1  and x2 based on the formula presented in the paper
  x1_new = x1_current + 0.03 * ( (x2_current * max( c2_current + p2 - c1_current - p1 , 0)) -
                                (x1_current * max( c1_current + p1 - c2_current - p2 , 0)) )
  
  x2_new = x2_current + 0.03 * ( (x1_current * max( c1_current + p1 - c2_current - p2 , 0)) -
                                (x2_current * max( c2_current + p2 - c1_current - p1 , 0)) )
  
  # Updating x1 and x2
  x1_current = x1_new
  x2_current = x2_new

  # Adding the new values for x1 and x2 to their defied lists
  x1.append(x1_current)
  x2.append(x2_current)

print(x1)
print(x2)
```

```python
# Plotting the graphs 
plt.plot(x1)
plt.plot(x2)
  
# Naming the x axis
plt.xlabel('n')
# Naming the y axis
plt.ylabel('Flow')
  
# Giving a title to my graph
plt.title('The evolutionary trajectories’ figure')
  
# Function to show the plot
plt.show()
```

```python
# First values for x1 and x2
x1_current = 1
x2_current = 5

# Defining lists for x1 and x2 to store the value of these two variables in
x1 = [x1_current]
x2 = [x2_current]

# Frist values for p1 and p2
p1_current = -4
p2_current = 4
```

```python

for i in range(2,101):

  # First, calculating c1 and c2 based on the current x1 and x2
  c1_current = x1_current + (x1_current * x2_current) 
  c2_current = x2_current + 1

  # Second, calculating p1 and p2
  p1_new = p1_current - 0.05*(x1_current - x2_current)
  p2_new = p2_current - 0.05*(x2_current - x1_current) 

  # Calculating x1  and x2 based on the formula presented in the paper and updated c1, c2, p1, p2
  x1_new = x1_current + 0.03 * ( (x2_current * max( c2_current + p2_new - c1_current - p1_new , 0)) -
                                (x1_current * max( c1_current + p1_new - c2_current - p2_new , 0)) )
  
  x2_new = x2_current + 0.03 * ( (x1_current * max( c1_current + p1_new - c2_current - p2_new , 0)) -
                                (x2_current * max( c2_current + p2_new - c1_current - p1_new , 0)) )
  
  # Updating x1 and x2
  x1_current = x1_new
  x2_current = x2_new

  # Adding the new values for x1 and x2 to their defied lists
  x1.append(x1_current)
  x2.append(x2_current)

  # Updating p1 and p2
  p1_current = p1_new
  p2_current = p2_new

print(x1)
print(x2)
```
